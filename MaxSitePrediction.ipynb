```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
preds = pd.read_csv("sitePredictions.csv")
preds = np.array(preds)
maxVals = []
for i in range(preds.shape[0]):
    maxVals.append(preds[i,1:-1].max())
```

```python
fig= plt.figure(figsize=(15, 5))
plt.ylim(0, 150)
plt.yticks([40, 80, 120])
plt.plot(maxVals)
```
